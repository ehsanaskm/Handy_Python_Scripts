```python
import numpy as np
import matplotlib.pyplot as plt

#data = np.loadtxt("Eau-2.4LM.csv",skiprows=1,delimiter=",")

A = np.loadtxt("data_mixer.csv",comments="#",delimiter=",",skiprows=1)

C = 0.13;
w = 0.026;
omega=53.12

uTip = 0.086/2*omega;


plt.plot(A[:,2]/uTip,(A[:,16]-C)/(w/2),'y-.')
plt.xlabel("$U_z/V_{tip}$")
plt.ylabel('2 z / w')
plt.grid(True)
plt.legend(loc=2,bbox_to_anchor=(0.6,0.5))
plt.savefig("Uz.eps", format="eps")
plt.show()


plt.plot(np.sqrt(A[:,0]**2+A[:,1]**2)/uTip, (A[:,16]-C)/(w/2), '*')
plt.xlabel("$U_r/V_{tip}$")
plt.ylabel('2 z / w')
plt.grid(True)
plt.legend(loc=2,bbox_to_anchor=(0.6,0.5))
plt.savefig("Ur.eps", format="eps")
plt.show()



```
