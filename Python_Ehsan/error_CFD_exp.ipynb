```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as intrp
from numpy import linalg as LA

%matplotlib inline

exp = np.loadtxt("exp.csv",skiprows=1,delimiter=",",dtype='float')
cfd = np.loadtxt("cfd.csv",skiprows=1,delimiter=",",dtype='float')

fi=intrp.interp1d(cfd[:,0],cfd[:,1],fill_value="extrapolate")


plt.plot(exp[:,0],fi(exp[:,0]),exp[:,0],exp[:,1],'*')
plt.rcParams["figure.figsize"] = 8,9

plt.grid()

ssv=LA.norm(exp[:,1]-fi(exp[:,0]))

ssv2=LA.norm(exp[:,1]*2*np.pi*exp[:,0]-fi(exp[:,0])*2*np.pi*exp[:,0]);

print(ssv)
print(ssv2)
```
